# Share Plots - Bar Charts

This notebook serves as a helpful guide for generating visually appealing figures using popular Python plotting libraries. Whether you're a researcher, data scientist, or student, well-crafted bar charts are essential for presenting data in a clear and impactful manner. In this notebook, you will find easy-to-follow code examples and step-by-step instructions to create various types of charts, including:
1. Single grouped bar chart
2. Multiple grouped bar chart

## Initialization

```python
# built-in
import os
# public
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 26})
colors = [plt.cm.Paired(i) for i in range(20)]
# private
from src import utils

%load_ext autoreload
%autoreload 2
%config Completer.use_jedi = False
```

---
## 1. Single grouped bar chart

```python
# init
data_path = os.path.join('res', 'bar', '1.pkl')
data_dict = utils.load_pickle(data_path)

n_tasks = len(data_dict['datasets'])
```

```python
# bar chart
fig, ax = plt.subplots(1, 1, sharey=True, figsize=(18, 6), dpi=80)

ax.bar(
    np.arange(n_tasks)-0.1
    , data_dict['roberta_pi']
    , width=0.2
    , color=colors[0]
    , edgecolor='black'
    , linewidth=3
    , label = 'RoBERTa PI'
)

ax.bar(
    np.arange(n_tasks)
    , data_dict['roberta']
    , width=0.2
    , color=colors[2]
    , edgecolor='black'
    , linewidth=3
    , label = 'RoBERTa NLI'
)

ax.bar(
    np.arange(n_tasks)+0.1
    , data_dict['xlnet']
    , width=0.2
    , color=colors[4]
    , edgecolor='black'
    , linewidth=3
    , label = 'XLNet NLI'
)

ax.set_xticks([0, 1, 2, 3, 4, 5])
ax.set_xticklabels(data_dict['datasets'], fontsize=24)

ax.set_ylim([70, 100])

ax.set_ylabel('F1 (%)', fontsize=30)
ax.set_xlabel('PI Datasets', fontsize=30)

ax.yaxis.grid(True, linestyle='--', linewidth=0.5, color='gray', alpha=0.5)


fig.legend(loc='upper center', bbox_to_anchor=(0.5, 1.12), ncol=5, fancybox=True, shadow=True, fontsize=30)

plt.tight_layout()
plt.show()
```

## 2. Multiple grouped bar chart

```python
# init
data_path = os.path.join('res', 'bar', '2.pkl')
data_dict = utils.load_pickle(data_path)

symm_roberta_0shot_pos_results = data_dict['symm_roberta_0shot_pos_results']
symm_roberta_0shot_neg_results = data_dict['symm_roberta_0shot_neg_results']

symm_roberta_ft_pos_results = data_dict['symm_roberta_ft_pos_results']
symm_roberta_ft_neg_results = data_dict['symm_roberta_ft_neg_results']

asym_roberta_0shot_pos_results = data_dict['asym_roberta_0shot_pos_results']
asym_roberta_0shot_neg_results = data_dict['asym_roberta_0shot_neg_results']

asym_roberta_ft_pos_results = data_dict['asym_roberta_ft_pos_results']
asym_roberta_ft_neg_results = data_dict['asym_roberta_ft_neg_results']

datasets = ['PIT', 'QQP', 'MRPC', 'PAWS QQP', 'PAWS Wiki', 'PARADE']
```

```python
fig, axs = plt.subplots(2, 2, sharey=True, figsize=(24, 6), dpi=80)

axs[0][0].bar(
    np.arange(len(datasets))-0.1
    , asym_roberta_0shot_pos_results
    , width=0.3
    , color=colors[4]
    , edgecolor='black'
    , linewidth=3
    , label = 'Positive Accuracy'
)

axs[0][0].bar(
    np.arange(len(datasets))+0.1
    , asym_roberta_0shot_neg_results
    , width=0.3
    , color=colors[0]
    , edgecolor='black'
    , linewidth=3
    , label = 'Negative Accuracy'
)

axs[1][0].bar(
    np.arange(len(datasets))-0.1
    , asym_roberta_ft_pos_results
    , width=0.3
    , color=colors[4]
    , edgecolor='black'
    , linewidth=3
)
axs[1][0].bar(
    np.arange(len(datasets))+0.1
    , asym_roberta_ft_neg_results
    , width=0.3
    , color=colors[0]
    , edgecolor='black'
    , linewidth=3
)

axs[0][1].bar(
    np.arange(len(datasets))-0.1
    , symm_roberta_0shot_pos_results
    , width=0.3
    , color=colors[4]
    , edgecolor='black'
    , linewidth=3
)
axs[0][1].bar(
    np.arange(len(datasets))+0.1
    , symm_roberta_0shot_neg_results
    , width=0.3
    , color=colors[0]
    , edgecolor='black'
    , linewidth=3
)

axs[1][1].bar(
    np.arange(len(datasets))-0.1
    , symm_roberta_ft_pos_results
    , width=0.3
    , color=colors[4]
    , edgecolor='black'
    , linewidth=3
)
axs[1][1].bar(
    np.arange(len(datasets))+0.1
    , symm_roberta_ft_neg_results
    , width=0.3
    , color=colors[0]
    , edgecolor='black'
    , linewidth=3
)

axs[0][0].set_xticks(range(len(datasets)))
axs[0][0].set_xticklabels(datasets, fontsize=20)
axs[0][1].set_xticks(range(len(datasets)))
axs[0][1].set_xticklabels(datasets, fontsize=20)
axs[1][0].set_xticks(range(len(datasets)))
axs[1][0].set_xticklabels(datasets, fontsize=20)
axs[1][1].set_xticks(range(len(datasets)))
axs[1][1].set_xticklabels(datasets, fontsize=20)

axs[0][0].set_yticks([0, 30, 60, 90])
axs[0][0].set_yticklabels([0, 30, 60, 90], fontsize=24)
axs[1][0].set_yticks([0, 30, 60, 90])
axs[1][0].set_yticklabels([0, 30, 60, 90], fontsize=24)

axs[0][0].yaxis.grid(True, linestyle='--', linewidth=1, color='gray', alpha=0.5)
axs[0][1].yaxis.grid(True, linestyle='--', linewidth=1, color='gray', alpha=0.5)
axs[1][0].yaxis.grid(True, linestyle='--', linewidth=1, color='gray', alpha=0.5)
axs[1][1].yaxis.grid(True, linestyle='--', linewidth=1, color='gray', alpha=0.5)

axs[0][0].set_title('(a) Aysm. Zero-shot', pad=15, fontsize=30)
axs[1][0].set_title('(c) Aysm. Fine-tuned', pad=15, fontsize=30)

axs[0][1].set_title('(b) Symm. Zero-shot', pad=15, fontsize=30)
axs[1][1].set_title('(d) Symm. Fine-tuned', pad=15, fontsize=30)

fig.legend(loc='upper center', bbox_to_anchor=(0.5, 1.12), ncol=2, fancybox=True, shadow=True, fontsize=30)

plt.tight_layout()
plt.show()
```
